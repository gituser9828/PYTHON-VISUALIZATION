```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
SS = pd.read_csv(r"C:\Users\SWAPNIL BANSODE\Downloads\Superstore.csv")
SS.info()
```

```python
plt.scatter(SS.Sales, SS.Profit, c = SS.Discount, cmap = 'summer', edgecolor = 'black',
            linewidth = 0.8)
cbar = plt.colorbar()
plt.xlabel('Sales')
plt.ylabel('Profit')
#cbar.set_label('Discount')
plt.show()
```

```python
plt.scatter(SS.Sales,SS.Profit,c=SS.Discount,cmap="summer",edgecolor="black",linewidth=0.8)
cbar=plt.colorbar()
cbar.set_label("Discount")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
plt.scatter(SS.Sales,SS.Profit,c=SS.Discount,cmap="summer",edgecolor="black",linewidth=0.8)
cbar=plt.colorbar()
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.xscale("log")
cbar.set_label("Discount")
```

```python
plt.hist(SS.Segment)
plt.show()
```

```python
#plot a bar graph showing total sales by each product category
categorey_wise_sales=SS.groupby("Category")["Sales"].sum()
```

```python
#categorey_wise_sales=SS.groupby("Category")["Sales"].sum()
plt.figure(figsize=(6,5))
categorey_wise_sales.plot(kind="bar")
plt.title("CATEGORY_WISE_SALES_")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
TOP_SELLING_PRODUCTS=SS.groupby("Product Name")["Sales"].sum().nlargest(10)
```

```python
plt.figure(figsize=(10,6))
TOP_SELLING_PRODUCTS.plot(kind="bar")
plt.title("RANKING OF PRODUCTS NY THEIR SALES")
plt.xlabel("Products Name")
plt.ylabel("Total Sales")
plt.xticks(rotation=75)
plt.show()
```
